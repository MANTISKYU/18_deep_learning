```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace

#   
img_path1 = './images/young1.jpg' 
img_path2 = './images/young3.jpg'

#  
img1 = cv2.imread(img_path1)
img2 = cv2.imread(img_path2)

#    
if img1 is None:
    print(f"{img_path1}   .")
if img2 is None:
    print(f"{img_path2}   .")

if img1 is not None and img2 is not None:
    plt.figure(figsize=(10, 5))  #   

    #    
    plt.subplot(1, 2, 1)  # 1 2   
    plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
    plt.title("Image 1") 
    plt.axis('off') 

    #    
    plt.subplot(1, 2, 2)  # 1 2   
    plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
    plt.title("Image 2")  
    plt.axis('off')  

    #  
    plt.show()

    #    
    try:
        result = DeepFace.verify(img_path1, img_path2, model_name='ArcFace')
        print(f"  : {result['verified']}")
    except Exception as e:
        print(f" : {e}")
        print(f"  1: {img_path1},   2: {img_path2}")
```
