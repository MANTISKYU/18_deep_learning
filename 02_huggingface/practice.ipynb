```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace

# 이미지 경로 설정
img_path1 = './images/young1.jpg' 
img_path2 = './images/young3.jpg'

# 이미지 읽기
img1 = cv2.imread(img_path1)
img2 = cv2.imread(img_path2)

# 이미지가 정상적으로 읽혔는지 확인
if img1 is None:
    print(f"{img_path1}를 찾을 수 없습니다.")
if img2 is None:
    print(f"{img_path2}를 찾을 수 없습니다.")

if img1 is not None and img2 is not None:
    plt.figure(figsize=(10, 5))  # 전체 크기 조정

    # 첫 번째 이미지 확인
    plt.subplot(1, 2, 1)  # 1행 2열의 첫 번째 위치
    plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
    plt.title("Image 1") 
    plt.axis('off') 

    # 두 번째 이미지 확인
    plt.subplot(1, 2, 2)  # 1행 2열의 두 번째 위치
    plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
    plt.title("Image 2")  
    plt.axis('off')  

    # 이미지 표시
    plt.show()

    # 동일 인물 여부 확인
    try:
        result = DeepFace.verify(img_path1, img_path2, model_name='ArcFace')
        print(f"동일 인물 여부: {result['verified']}")
    except Exception as e:
        print(f"오류 발생: {e}")
        print(f"이미지 경로 1: {img_path1}, 이미지 경로 2: {img_path2}")
```
